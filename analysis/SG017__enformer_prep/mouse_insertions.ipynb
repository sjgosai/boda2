```python
oligo_keys = []
enfo_maxs  = np.zeros(shape=(0,6956))
enfo_means = np.zeros(shape=(0,6956))

for i in tqnb.tqdm(range(1,100)):
    !gsutil cp -n gs://korvaz/mpra_model_manuscript/data/enformer_mouse_insert_preds/lacz/chunk_{i}.npy ./
    check = np.load(f'chunk_{i}.npy', allow_pickle=True)
    keys_, preds_ = zip(*[ (k, v) for k, v in check.item().items() if k != 'blank' ])
    preds_ = np.squeeze(np.concatenate(preds_, axis=0) - check.item()['blank'])
    oligo_keys.extend(keys_)
    enfo_maxs  = np.concatenate([enfo_maxs,   np.max(preds_, axis=1)], axis=0)
    enfo_means = np.concatenate([enfo_means, np.mean(preds_, axis=1)], axis=0)
    !rm chunk_{i}.npy
```

```python
df_maxes = pd.DataFrame(enfo_maxs, index=oligo_keys)
df_means = pd.DataFrame(enfo_means, index=oligo_keys)
```
